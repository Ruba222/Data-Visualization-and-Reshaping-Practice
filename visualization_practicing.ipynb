```python
import pandas as pd
import numpy as np 
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
days = ['Mon','Tue','Wed','Thu','Fri','Sat','Sun']
temp_cairo = [22.5,22.7,25,27.3,26.1,22.4,20]
temp_aswan = [30.2,29,25,33.3,39.7,35,28]
```

```python
cairo = pd.DataFrame(data = {'Day' : days , 'Tempratures' :temp_cairo })
cairo
```

```python
aswan = pd.DataFrame(data = {'Day' : days , 'Tempratures' :temp_aswan})
aswan
```

```python
plt.plot(cairo['Day'],cairo['Tempratures'])
```

```python
plt.scatter(cairo['Day'],cairo['Tempratures'])
```

```python
plt.bar(cairo['Day'],cairo['Tempratures'])
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [10, 20, 15, 25]

plt.plot(x, y, color='red', linestyle=':', marker='o', markersize=10, label='Data')

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Example Plot")
plt.legend()
plt.show()
```

```python
plt.plot(cairo['Day'],cairo['Tempratures'], ls = ':', lw = 2, c = 'red', marker = 'o', mec = 'blue', label = 'Cairo', alpha = 0.9,  markersize = 8)
plt.plot(aswan['Day'],aswan['Tempratures'], ls = '-', lw = 2, c = 'blue', marker = '*', mec = 'red', label = 'Aswan', alpha = 0.7,  markersize = 8)

plt.ylabel("tempratures in cel")
plt.xlabel('Days of the week')
plt.title('Temprature Degrees of Cairo & Aswan')
plt.legend(loc = 0)
#plt.ylim(0,100)
```

```python
figure,axes = plt.subplots(1,2,figsize = (10,5))
axes[0].plot(cairo['Day'],cairo['Tempratures'],c = 'k')
axes[0].set_title('Tempratures degrees in Aswan')


axes[1].plot(aswan['Day'],aswan['Tempratures'],c = 'blue')
axes[1].set_title('temp of aswan')
```

```python
x = np.linspace(1,10,11)
y = x**2
x
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.7,0.7])
axes1.set_title('axes_1')

axes2 = fig.add_axes([0.2,0.5,0.2,0.2])
axes2.set_title('axes_2')

axes1.plot(x,y,'b')
axes2.plot(y,x,'k')
```
