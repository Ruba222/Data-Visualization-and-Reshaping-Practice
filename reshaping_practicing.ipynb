```python
titanic.pivot_table(index='sex',columns='class',values=('fare','survived'),margins=True,aggfunc='sum').reset_index()
```

```python
titanic.pivot_table(index='sex',columns='class' ,margins=True).reset_index()
```

```python
titanic.pivot_table(index=['sex','pclass'],values = ('age','survived','fare'))
```

```python
#aggregation function can be like this (i can specify how i want each column to be agg)

titanic.pivot_table(index=['sex','pclass'],values = ('age','survived','fare'),aggfunc = {'age':np.mean,'survived':np.sum,'fare':np.max})
```

```python
#also i can do more than one agg function on the same column
titanic.pivot_table(index=['sex','pclass'],values = 'age' ,aggfunc=[np.sum,np.mean])
```

```python
titanic.pivot_table(index='sex', columns = 'pclass' ,values = 'age' ,aggfunc=[np.sum,np.mean])
```

```python
# now we can fill null values by using fill_value()
pd.pivot_table(titanic,
               index=['sex','survived','pclass'],
              columns = ['embark_town'],
              values = ['age'],
               aggfunc=np.mean,
               fill_value=titanic['age'].mean(),
               margins=True
              ).reset_index()
```

```python
df = pd.DataFrame({'first':{0:'jhon',1:'mary'},
                  'last':{0:'Doe',1:'Bo'},
                  'height':{0:5.5,1:6.0},
                  'weight':{0:130,1:150}})
df
```

```python
pd.melt(df,id_vars=('first','last'),
           var_name='h&w',
           value_name ='h&w_values').reset_index()
```
