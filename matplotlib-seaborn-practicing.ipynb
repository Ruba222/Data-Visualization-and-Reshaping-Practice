Create a figure object and put two axes on it, ax1 and ax2. Located at [0,0,1,1] and [0.2,0.5,.2,.2] respectively.

```python
n =  [0,0.2,0.4,0.6,0.8,1]
m = [0,0.2,0.4,0.6,0.8,1]
fig = plt.figure();
ax1 = fig.add_axes([0,0,1,1]);
ax2 = fig.add_axes([0.2,0.5,.2,.2])
ax1.plot(n,m)
ax2.plot(m,n)
ax2.set_xlim(min(m), max(m))
ax2.set_ylim(min(n), max(n))
```

```python
x1 = [0, 20, 40, 60, 80, 100]
y1 = [0, 50, 100, 150, 200,250]

fig = plt.figure()
ax1 = fig.add_axes([0, 0, 1, 1])
ax2 = fig.add_axes([0.2, 0.5, 0.2, 0.2])

ax1.plot(x1, y1)
ax2.plot(x1, y1)

# Corrected method names
ax1.set_xlabel("x")
ax1.set_ylabel("y")

ax2.set_xlabel("x")
ax2.set_ylabel("y")




ax2.set_xlim(0,100)
ax2.set_ylim(min(y1), max(y1))
```

```python
figure, axes = plt.subplots(1,2,figsize=(5,5))
axes[0].plot(m,m)
axes[1].plot(m,n)
```

# Seaborn Exercises

## Titanic Data

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

#### Recreate the plots below using the titanic dataframe.

```python
sns.jointplot(x = 'fare',y='age',data=titanic, kind = 'scatter')
```

```python
titanic[["fare"]].plot.hist(bins=50)
```

```python
import seaborn as sns 
sns.set()
%matplotlib inline

sns.boxplot(x='class',y='age',data=titanic,palette = 'Set3')
```

```python
sns.countplot(x='sex',data=titanic)
```
